```
import numpy as np
import matplotlib.pyplot as plt
```

```
img=plt.imread('/content/download.jpg')
```

```
plt.imshow(img)
```

```
img.shape
```

```
# using opencv library for face detection
# using cascade classifiers
# haar cascade is are machine learning models in open cv
# go to opencv github and select haarcascade_frontalface_default.xml
# this is an xml file
# https://github.com/opencv/opencv/blob/master/data/haarcascades/haarcascade_frontalface_default.xml
```

```
import cv2
# to access the xml file
model=cv2.CascadeClassifier('/content/haarcascade_frontalface_alt.xml')
```

```
# a function in 
faces=model.detectMultiScale(img)
faces
# this will return an numpy array
# x-coordinate, y-coordinate, width, height of rectangle to plot
# this rectangle has the face that is detected
```

```
faces.shape
# shape is number of rows(number of faces) and number of columns
```

```
for face in faces:
  x,y,w,h=face
  # img,(top left coordiinate),(bottom right coordinate),(color in rgb),(width og box)
  img=cv2.rectangle(img,(x,y),(x+w,y+h),(255,255,255),2)
```

```
plt.imshow(img)
```
